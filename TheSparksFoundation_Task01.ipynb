**Predicting the percentage of Student on the basis of numbers of hours studied**

## Importing the Necessary Headers

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
%matplotlib inline
```

## Reading the Csv files And Analysing the data

```python
data=pd.read_csv('http://bit.ly/w-data')
```

```python
data.iloc[0:5] 
# to visualise the basics of data
```

```python
plt.scatter(data.Hours,data.Scores)
plt.xlabel("Number of Hours studied")
plt.ylabel("Marks Achieved")
plt.grid()
```

**From the Above Scatter-plot We analysed the Linear-Like Relationship between the number of Hours studied and the Marks Obtained.** 

```python
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values
```

```python
X[0:5]
```

```python
y[0:5]
```

##Splitting the Data into Training and Test(80-20 rule)

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
regressor=LinearRegression(fit_intercept=True)
regressor.fit(X_train,y_train)
```

```python
print("Equation of Best fit line is: {:.2f}x +{:.2f}".format(regressor.coef_[0], regressor.intercept_))
```

```python
line=regressor.coef_*X+regressor.intercept_
```

```python
plt.scatter(X, y)
plt.title('Best Fit Line')
plt.xlabel('Number of hours studied')
plt.ylabel('Percentage of Marks obtained')
plt.plot(X, line,'green')
plt.grid()
```

```python
print(X_test)
y_pred = regressor.predict(X_test)
```

```python
hours = np.array(9.25)
hours=hours.reshape(-1,1)
own_pred = regressor.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

## Evaluating the Model Mathematically

```python
print("Score of the model is ",regressor.score(X_test,y_test))
print("Mean square Error is ",metrics.mean_squared_error(y_test,y_pred))
print("Root Mean square error is",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
